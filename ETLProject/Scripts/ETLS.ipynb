```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop'
browser.visit(url)
```

```python
resultados = []

for x in range(20):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    quotes = soup.find_all('b')
    names = soup.find_all('td', class_='hauptlink')
    for name in names:
        resultados.append(name.text)
    browser.visit(f'https://www.transfermarkt.com/spieler-statistik/wertvollstespieler/marktwertetop?ajax=yw1&page={x + 2}')
```

```python
print(resultados[0])
```

```python
jugadores = []
valor = []

for x in range(len(resultados)):
    if x%2 == 0:
        jugadores.append(resultados[x])
    if x%2 == 1:
        valor.append(resultados[x])
```
